# There are 3 rules in broadcasting.</br>
**Rule 1, says:** If the two arrays differ in their number of dimensions, then the shape of the one with fewer dimensions is padded with ones on its leading left side. </br>
**Rule 2 says:** If the shape of the two arrays does not match in any dimension, the array with shape equal to 1 in that dimension is stretched to match the other shape. </br>
**Rule 3 says:** If in any dimension, the sizes disagree and neither is equal to 1, an error is raised. </br>
That means, Broadcasting is not possible.

```python
x = np.arange(15).reshape(5,3)
y = np.arange(3)
print(x.shape)
print(y.shape)
```

```python
result = x + y
print(result.shape)
```
